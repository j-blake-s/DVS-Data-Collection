```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with np.load("ASL-Signs.npz") as data:
  trainX = data['tX']
  trainY = data['tY']
  valX = data['vX']
  valY = data['vY']
```

```python
import torch
import torch.nn as nn
layer = nn.MaxPool3d((2,2,2), (2,2,2))
temp = torch.from_numpy(trainX)
temp = torch.permute(temp, [0,2,3,4,1])

out = layer(temp)
```
